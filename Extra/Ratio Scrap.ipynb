```python
# Find the <h1> element with the specific class
h1_element = soup.find('h1', class_='js-recent-quote-ticker')

# Extract the ticker from the data-ticker attribute
if h1_element:
    ticker = h1_element.get('data-ticker')
    print("Ticker:", ticker)
```

```python
t_dic = {
    'ticker' : [],
    'ROA' : [], 
    'ROE' : [], 
    'ROI' : [], 
    'Gross Margin' : [], 
    'Oper. Margin' : [], 
    'Profit Margin' : []
}
```

```python
for url in tqdm(urls):
    response = requests.get(url,headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the <h1> element with the specific class
    h1_element = soup.find('h1', class_='js-recent-quote-ticker')
    
    # Extract the ticker from the data-ticker attribute
    if h1_element:
        ticker = h1_element.get('data-ticker')
    t_dic['ticker'].append(ticker)

        # Find the <td> element containing the quick ratio
    ROA_td = soup.find('td', string ='ROA')
    
    # Check if the <td> element is found
    if ROA_td:
        # Get the next <td> sibling element which contains the quick ratio value
        ROA_value = ROA_td.find_next_sibling('td').text.strip()
    t_dic['ROA'].append(ROA_value)
    
    # Find the <td> element containing the quick ratio
    ROE_td = soup.find('td', string ='ROE')
    
    # Check if the <td> element is found
    if ROE_td:
        # Get the next <td> sibling element which contains the quick ratio value
        ROE_value = ROE_td.find_next_sibling('td').text.strip()
    t_dic['ROE'].append(ROE_value)
    
    # Find the <td> element containing the quick ratio
    ROI_td = soup.find('td', string ='ROI')
    
    # Check if the <td> element is found
    if ROI_td:
        # Get the next <td> sibling element which contains the quick ratio value
        ROI_value = ROI_td.find_next_sibling('td').text.strip()
    t_dic['ROI'].append(ROI_value)
            
    # Find the <td> element containing the Gross Margin
    gross_margin_td = soup.find('td', string='Gross Margin')
    
    # Check if the <td> element is found
    if gross_margin_td:
        # Get the next <td> sibling element which contains the Gross Margin value
        gross_margin_value = gross_margin_td.find_next_sibling('td').text.strip()
    t_dic['Gross Margin'].append(gross_margin_value)
    
    # Find the <td> element containing the Operating Margin
    oper_margin_td = soup.find('td', string='Oper. Margin')
    if oper_margin_td:
        oper_margin_value = oper_margin_td.find_next_sibling('td').text.strip()
    t_dic['Oper. Margin'].append(oper_margin_value)
    
    # Find the <td> element containing the Profit Margin
    profit_margin_td = soup.find('td', string='Profit Margin')
    if profit_margin_td:
        profit_margin_value = profit_margin_td.find_next_sibling('td').text.strip()
    t_dic['Profit Margin'].append(profit_margin_value)
```
